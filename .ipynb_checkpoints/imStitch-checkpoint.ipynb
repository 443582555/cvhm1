```python
%matplotlib inline
import cv2
from pano import pano
```

```python
import sys
import numpy as np
import glob
from matplotlib import pyplot as plt
```

```python
if __name__ == '__main__':
    sourceImg=[]
    imdir = '/img'
    for fn in sorted(glob.glob('img/*.JPG')):
        im = cv2.imread(fn)
        sourceImg.append(im)
```

```python
fig = plt.figure()
size=len(sourceImg)
plt.imshow(sourceImg[1])
```

```python
mstich = pano(sourceImg[1],sourceImg[2])
```

```python
kp1,kp2,good = mstich.xsiftFeature()
```

```python
M = mstich.get_homography(kp1,kp2,good)
```

```python
result = mstich.get_stitched_image(M)
```

```python
fig = plt.figure()
size=len(sourceImg)
plt.imshow(result)
```

```python
currentImg = sourceImg[0]
for i in range(10):
        mstich = pano(currentImg,sourceImg[i+1])
        kp1,kp2,good = mstich.xsiftFeature()
        M = mstich.get_homography(kp1,kp2,good)
        result = mstich.get_stitched_image(M)
        currentImg = result
fig = plt.figure()
size=len(sourceImg)
plt.imshow(result)
```
